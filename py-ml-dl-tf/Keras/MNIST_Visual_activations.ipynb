
<h1 id="Classification">Classification<a class="anchor-link" href="#Classification">¶</a></h1>


```python

random.seed(1234)
train_proc = 0.75
data_train, data_test = sklearn.model_selection.train_test_split(data, test_size = 0.25)

x_train = data_train.iloc[:, 1:]
y_train = data_train.iloc[:, 0]
x_test = data_test.iloc[:, 1:]
y_test = data_test.iloc[:, 0]

x_train = np.array(x_train)
y_train = np.array(y_train)
x_test = np.array(x_test)
y_test = np.array(y_test)
```


<h2 id="Logistic-regression">Logistic regression<a class="anchor-link" href="#Logistic-regression">¶</a></h2><h3 id="Model">Model<a class="anchor-link" href="#Model">¶</a></h3>


```python

model_lr = sklearn.linear_model.LogisticRegression(max_iter = 100, 
                                                   multi_class = "multinomial", 
                                                   solver = "lbfgs", verbose = 3).fit(x_train, y_train)
```


<h3 id="Predict">Predict<a class="anchor-link" href="#Predict">¶</a></h3>


## Keras Deep Neural Net
### Prepare

```python

print("x_train shape", x_train.shape)
print("y_train shape", y_train.shape)
print("x_test shape", x_test.shape)
print("y_test shape", y_test.shape)
```


<p>Converting a Pandas DataFram to a Numpy Array before has already reshaped the array.</p>


```python

# reshape
# x_train = x_train.reshape(x_train.shape[0], 784)
# x_test = x_test.reshape(x_test.shape[0], 784)

# rescale
x_train = x_train.astype("float32")
x_test = x_test.astype("float32")
x_train /= 255
x_test /= 255

# one hot encode
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
```

```python

print("Train matrix shape", x_train.shape)
print("Test matrix shape", x_test.shape)

print("y_train shape after one-hot encoding: ", y_train.shape)
```


<h3 id="Model">Model<a class="anchor-link" href="#Model">¶</a></h3>


```python

model_ker = Sequential()
model_ker.add(Dense(units = 256, activation = "relu", input_shape = (784, )))
model_ker.add(Dropout(rate = 0.4))
model_ker.add(Dense(units = 64, activation = "relu"))
model_ker.add(Dropout(rate = 0.3))
model_ker.add(Dense(units = 10, activation = "softmax"))

model_ker.summary()
```

```python

model_ker.compile(loss = keras.losses.categorical_crossentropy,
                 optimizer = keras.optimizers.RMSprop(),
                 metrics = ["accuracy"])
```


<h3 id="Train">Train<a class="anchor-link" href="#Train">¶</a></h3>


```python
history = model_ker.fit(x_train, y_train,
             epochs = 30,
             batch_size = 100,
             validation_data = [x_test, y_test])
```

```python
plt.plot(history.history["acc"])
plt.plot(history.history["val_acc"])
plt.title("Accuracy")
plt.ylabel("accuracy")
plt.xlabel("epoch")
```

### Save model and reload it

```python
# Save the mode made by DNN
model_json = model_ker.to_json()
with open("model_dnn.json", "w") as json_file : 
    json_file.write(model_json)

model_ker.save_weights("model_dnn.h5")
print("Saved model to disk")
```

```python
# 저장된 JSON 파일로 부터 모델 로드하기

from keras.models import model_from_json
json_file = open("model_dnn.json", "r")
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)


# 로드한 모델에 Weight 로드하기

loaded_model.load_weights("model_dnn.h5")
print("Loaded model from disk")


# 모델 컴파일 후 Evaluation

loaded_model.compile(loss="binary_crossentropy", optimizer="rmsprop", metrics=['accuracy'])

# model evaluation
score = loaded_model.evaluate(x_test, y_test,verbose=0)

print("%s : %.2f%%" % (loaded_model.metrics_names[1], score[1]*100))


# 출처: https://3months.tistory.com/150 [Deep Play]
```

## Keras CNN
### Prepare
> Data is already rescaled and encoded. We only need to reshape it as a square image of 28 X 28 X 1 pixels.

```python

# reshape
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
```

```python

print("Train matrix shape", x_train.shape)
print("Test matrix shape", x_test.shape)

print("y_train shape after one-hot encoding: ", y_train.shape)
print("Max value in test data set: ", np.amax(x_test))

```


<h3 id="Model">Model<a class="anchor-link" href="#Model">¶</a></h3>


```python

model_cnn = Sequential()
model_cnn.add(Conv2D(filters = 20, kernel_size = (5, 5),
                     activation = "relu",
                     input_shape = (28, 28, 1)))
model_cnn.add(MaxPooling2D(strides = (2, 2), pool_size = (2, 2)))
model_cnn.add(Conv2D(filters = 50, kernel_size = (5, 5),
                    activation = "relu"))
model_cnn.add(MaxPooling2D(strides = (2, 2), pool_size = (2, 2)))
model_cnn.add(Flatten())
model_cnn.add(Dense(units = 500, activation = "relu"))
model_cnn.add(Dense(units = 10, activation = "softmax"))

model_cnn.summary()
```

```python

model_cnn.compile(loss = keras.losses.categorical_crossentropy,
                 optimizer = keras.optimizers.adadelta(lr = 0.05),
                 metrics = ["accuracy"])
```


## Train


```python

history = model_cnn.fit(x_train, y_train,
             epochs = 30,
             batch_size = 100,
             validation_data = [x_test, y_test])

plt.plot(history.history["acc"])
plt.plot(history.history["val_acc"])
plt.title("Accuracy")
plt.ylabel("accuracy")
plt.xlabel("epoch")
```

```python
# Save the mode made by CNN
model_json = model_cnn.to_json()
with open("model_cnn.json", "w") as json_file : 
    json_file.write(model_json)

model_cnn.save_weights("model_cnn.h5")
print("Saved model to disk")
```

```python
# visualize activations

def vis_acti(n):
    # prepare data
    x_act = data.loc[data["label"] == n]
    x_act = x_act.iloc[:, 1:]
    x_act = np.array(x_act)
#     x_act = np.array(x_act)
    x_act = x_act.reshape(x_act.shape[0], 28, 28, 1)
    x_act = x_act.astype("float32")
    x_act /= 255
    
    # Extracts the outputs of the top 4 layers
    layer_outputs = [layer.output for layer in model_cnn.layers[:4]] 
    
    # Creates a model that will return these outputs, given the model input
    activation_model = models.Model(inputs=model_cnn.input, outputs=layer_outputs) 
    
    activations = activation_model.predict(x_act) 
    layer_features = [act.shape[-1] for act in activations] # numbers of features(neurons) in all layers
    
    # large img
    first_layer_activation = activations[0]  # 1st CNN2d layer
    print(layer_outputs[0].name)
    print(first_layer_activation.shape)
    plt.matshow(first_layer_activation[0, :, :, 4], cmap='viridis')  # first_layer_activation at neuron-5
    
    layer_names = []
    for layer in model_cnn.layers[:4]:
        layer_names.append(layer.name) # Names of the layers, so you can have them as part of your plot

#     images_per_row = 10  # greatest common divisor of numbers of features(neurons) in all layers
    images_per_row = np.gcd.reduce(layer_features)

    for layer_name, layer_activation in zip(layer_names, activations): # Displays the feature maps
        n_features = layer_activation.shape[-1] # Number of features(neurons) in the feature map
        size = layer_activation.shape[1] #The feature map has shape (1, size, size, n_features).
        n_cols = n_features // images_per_row # Tiles the activation channels in this matrix
        display_grid = np.zeros((size * n_cols, images_per_row * size))
        for col in range(n_cols): # Tiles each filter into a big horizontal grid
            for row in range(images_per_row):
                channel_image = layer_activation[0,  # first data of the chosen number
                                                 :, :,
                                                 col * images_per_row + row]
                channel_image -= channel_image.mean() # Post-processes the feature to make it visually palatable
                channel_image /= channel_image.std()
                channel_image *= 64
                channel_image += 128
                channel_image = np.clip(channel_image, 0, 255).astype('uint8')
                display_grid[col * size : (col + 1) * size, # Displays the grid
                             row * size : (row + 1) * size] = channel_image
        scale = 1. / size
        plt.figure(figsize=(scale * display_grid.shape[1],
                            scale * display_grid.shape[0]))
        plt.title(layer_name)
        plt.grid(False)
        plt.imshow(display_grid, aspect='auto', cmap='viridis')
        print(display_grid.shape,n_features,images_per_row,n_cols)
```
